```python
import pandas as pd
import numpy as np
df_data=pd.read_csv(r"D:\Mobadra\Preprocessing\googleplaystore.csv")

def convert_to_kb(Size):
    if 'M' in Size:
        return float(Size.replace('M',' ')) 
    elif 'k' in Size:
        return float(Size.replace('k',''))
    else:
        return None
df_data['Updated_Size'] = df_data['Size'].apply(convert_to_kb)
df_data
```

```python
df_data.dtypes
```

```python
df_data['Last Updated'] = pd.to_datetime(df_data['Last Updated'], errors='coerce')
df_data.dtypes
```

```python
def convert_price(Price):
    if pd.isna(Price):  
        return None  
    if isinstance(Price, str) and '$' in Price:  
        return float(Price.replace('$', ''))
    if isinstance(Price, (int, float)):  
        return float(Price)
    return None  
df_data['New_price'] = df_data['Price'].apply(convert_price)
print(df_data['New_price'].isna().sum())
df_data['New_price'] = df_data['New_price'].fillna(0)
print([df_data['New_price'].isna()])
```

```python
df_data['New_price'].sample(10)
```

```python
# df_data['Type'].unique()
# df_data['Type'].isnull().sum()
df_data['Type']=df_data['Type'].fillna("free")
df_data['Type'].isnull().sum()
df_data.isna().sum()
```

```python
df_data['Content Rating'].value_counts()
df_data['Content Rating']=df_data['Content Rating'].fillna("Everyone")

df_data.isna().sum()
```

```python
df_data['Current Ver']=df_data['Current Ver'].fillna("Varies with device")
df_data.isna().sum()
```

```python
df_data['Android Ver'].value_counts()
df_data['Android Ver']=df_data['Android Ver'].fillna("4.1 and up")
df_data.isna().sum()
```

```python
df_data.dtypes
```

```python
most_expensive_app = df_data[df_data['New_price'] == df_data['New_price'].max()]
print(most_expensive_app[['App', 'New_price']])
```

```python
genre_counts = df_data['Category'].value_counts()
most_common_genre = genre_counts.idxmax()
most_common_count = genre_counts.max()

print(f"{most_common_genre}' with {most_common_count} apps.")
```

```python
free_apps = df_data[df_data['New_price'] == 0]
paid_apps = df_data[df_data['New_price'] > 0]
avg_size_free = free_apps['Updated_Size'].mean()
avg_size_paid = paid_apps['Updated_Size'].mean()

print(f"Average size of free apps: {avg_size_free:.2f} MB")
print(f"Average size of paid apps: {avg_size_paid:.2f} MB")
```

```python
perfect_rated_apps = df_data[df_data['Rating']==5]
most_expensivet_apps = perfect_rated_apps.sort_values(by='New_price', ascending=False).head(5)

print(most_expensivet_apps[['App','Rating','New_price']])
```

```python
num_apps = df_data[df_data['Reviews'] > 50000].shape[0]

print(f"Number of apps with more than 50K reviews: {num_apps}")
```

```python
df_data['Installs'] = df_data['Installs'].str.replace('[+,]','',regex=True).astype(float)
avg_price = df_data.groupby(['Category', 'Installs'])['New_price'].mean().reset_index()
avg_price.head(10)
```

```python
high_rating_apps = df_data[df_data['Rating'] > 4.7]
num_of_rating_apps = high_rating_apps['Rating'].shape[0]
avg_of_rating_apps = high_rating_apps['New_price'].mean()
print(f"number of high ratinhgapps is: {num_of_rating_apps}")
print(f"average of high rating apps is: {avg_of_rating_apps:.2f}")
```

```python
df_data['Installs'] = df_data['Installs'].str.replace('[+,]','',regex=True)
df_data['Installs'] = df_data['Installs'].replace('Free',np.nan)
df_data['Installs'].value_counts()
```

```python
df_data['Installs'] = df_data['Installs'].str.replace('[+,]','',regex=True).astype(float)
High_install_apps = df_data[df_data['Installs'] >= 5000000]
app_sales = (df_data['Installs'] * High_install_apps['New_price']).sum()
google_revenue = app_sales * 0.30
print(f"Google's 30% cut is: {google_revenue:.2f}")
```

```python
df_data['Updated_Size'] = pd.to_numeric(df_data['Updated_Size'], errors='coerce')
df_data['App_Type'] = df_data['New_price'].apply(lambda x: 'Paid' if x > 0 else 'Free')
size_stats = df_data.groupby('App_Type')['Updated_Size'].agg(['min', 'max'])
size_stats
```
